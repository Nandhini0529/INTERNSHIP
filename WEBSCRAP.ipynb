```python
driver = webdriver.Chrome()

driver.get("https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos")
```

```python
url=[]

title_url=driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"]//tbody//tr//td[2]')

for i in title_url:
    url.append(i.text)
    
```

```python
rank=[]
ranks=driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"]//tbody//tr//td[4]')
for i in ranks:
    rank.append(i.text)
print(len(rank))    
    
```

```python
publisher=[]
publish=driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"]//tbody//tr//td[5]')
for i in publish:
    publisher.append(i.text)

print(len(publisher))    
```

```python
import pandas as pd
pdf=pd.DataFrame({"Video_name":url,"Uploader":name,"Views":rank,"Publication_date":publisher})
pdf
```
